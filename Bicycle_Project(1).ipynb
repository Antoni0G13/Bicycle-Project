```python
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance in kilometers between two points
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    km = 6371 * c  # Radius of earth in kilometers. Use 3956 for miles. Determines return value units.
    return km

# Calculate the distance using the Haversine formula
concatenated_df['distance'] = concatenated_df.apply(
    lambda row: haversine(row['start_lat'], row['start_lng'], row['end_lat'], row['end_lng']), axis=1
)

# Display the first few rows of the DataFrame with the new 'distance' column
concatenated_df[['start_lat', 'start_lng', 'end_lat', 'end_lng', 'distance']].head()
```

### Data Transoformation

```python
# Create a new column called 'trip_type' with initial value 'None'
concatenated_df['trip_type'] = None

# Display the DataFrame with the new column
concatenated_df.head()

# Explanation:
# The new column 'trip_type' has been created with an initial value of 'None'.
# This column will be used to estimate the type of trip based on the duration and distance of the trip.
# Different thresholds for trip duration and distance can be defined to classify trips into categories such as short trips, medium-distance trips, long trips, etc.
# By categorizing trips based on their characteristics, we can gain insights into different usage patterns and behaviors of bike-share users.
```

```python
# Define the trip type thresholds
duration_thresholds = {
    "short_trip": pd.Timedelta(minutes=5),
    "normal_trip": pd.Timedelta(minutes=15),
    "long_trip": pd.Timedelta(minutes=30),
}

distance_thresholds = {
    "short_trip": 1,   # in km
    "normal_trip": 5,  # in km
    "long_trip": 10,   # in km
}

# Function to classify trips
def classify_trip(row):
    if row['time_duration'] < duration_thresholds['short_trip'] and row['distance'] < distance_thresholds['short_trip']:
        return 'short_trip'
    elif row['time_duration'] < duration_thresholds['normal_trip'] and row['distance'] < distance_thresholds['normal_trip']:
        return 'normal_trip'
    elif row['time_duration'] < duration_thresholds['long_trip'] and row['distance'] < distance_thresholds['long_trip']:
        return 'long_trip'
    else:
        return 'very_long_trip'

# Apply the classification function to each row
concatenated_df['trip_type'] = concatenated_df.apply(classify_trip, axis=1)

# Display the DataFrame with the new column
print(concatenated_df[['started_at', 'ended_at', 'time_duration', 'distance', 'trip_type']].head())

# Explanation:
# The 'trip_type' column has been populated based on the defined thresholds for trip duration and distance.
# This classification helps in categorizing the trips into 'short_trip', 'normal_trip', 'long_trip', and 'very_long_trip'.
# By analyzing these categories, we can better understand the patterns and behaviors of bike-share users.
```

```python
import numpy as np

# Add the 'average_speed' column

# Ensure we avoid division by zero and handle cases where time duration is zero or negative
concatenated_df['average_speed'] = np.where(concatenated_df['time_duration'].dt.total_seconds() > 0,
                                            concatenated_df['distance'] / (concatenated_df['time_duration'].dt.total_seconds() / 3600),
                                            np.nan)  # km/h or NaN if time duration is zero or negative

# Add the 'trip_duration_minutes' column
concatenated_df['trip_duration_minutes'] = concatenated_df['time_duration'].dt.total_seconds() / 60  # minutes

# Justification:
# - Average Speed: Helps to analyze the efficiency and nature of trips, which is crucial for understanding transportation patterns and improving infrastructure.
# - Trip Duration in Minutes: Simplifies the interpretation and analysis of trip lengths, making it easier to derive actionable insights.
```
